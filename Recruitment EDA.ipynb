# Data Visulaisations

```python
# Visualizing the distribution of numeric columns
df.hist(figsize=(10,10))
plt.show()
```

```python
# Visualizing the correlation matrix
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.show()
```

```python
# Visualizing the count of categorical columns
sns.countplot(x='attrition', data=df)
plt.show()
```

```python
sns.countplot(x='performance_rating', data=df)
plt.show()
```

```python
sns.countplot(x='recruiting_source', data=df)
plt.show()
```

```python
# Visualizing the relationship between numeric and categorical columns
sns.boxplot(x='attrition', y='sales_quota_pct', data=df)
plt.show()
```

```python
sns.boxplot(x='performance_rating', y='sales_quota_pct', data=df)
plt.show()
```

```python
sns.boxplot(x='recruiting_source', y='sales_quota_pct', data=df)
plt.show()
```

```python
# plot a chart ggplot in python that illustrates attrition numbers by the recruiting sources
ggplot(df, aes(x='recruiting_source', fill='attrition')) + \
    geom_bar(position='dodge') + \
    labs(title="Attrition Numbers by Recruiting Sources", x="Recruiting Sources", y="Number of Employees") + \
    scale_fill_manual(values=["#4C72B0", "#C44E52"]) + \
    theme_minimal()
```

```python
# printout the average sales numbers grouped by recruiting source
avg_sales_by_source = df.groupby('recruiting_source')['sales_quota_pct'].mean()
avg_sales_by_source 
```

```python
# printout the average attrition numbers grouped by recruiting source
average_attrition = df.groupby('recruiting_source')['attrition'].mean()
average_attrition
```
